# Dropout
Dropout [1] is a technique for regularizing neural networks by randomly setting some output activations to zero during the forward pass. In this exercise, you will implement a dropout layer and modify your fully connected network to optionally use dropout.

[1] [Geoffrey E. Hinton et al, "Improving neural networks by preventing co-adaptation of feature detectors", arXiv 2012](https://arxiv.org/abs/1207.0580)

## Inline Question 1:
What happens if we do not divide the values being passed through inverse dropout by `p` in the dropout layer? Why does that happen?

## Answer:
If we do not divide the values by p in the dropout layer, we would encounter issues during the test phase.

During training, inverted dropout randomly sets a fraction p of the neuron's outputs to zero. This dropout operation helps to regularize the network by preventing overfitting and encourages robust learning. However, during the test phase, we want all neurons to contribute to the network's predictions because we're no longer in a learning phase. All neurons should see all their inputs.

If we don't scale the values by p during dropout at training time, we would face problems during testing. This is because at test time, all neurons should be active, and their outputs should be identical to what they would be in expectation during training.

Not dividing by p at training time means that the network's output values can become much larger than they should be. This can lead to issues like exploding gradients during backpropagation because the scale of the activations may have changed dramatically between training and testing.

To avoid this problem and for consistent behavior between training and testing, we use inverted dropout. Inverted dropout scales the values at training time, and this scaling ensures that during testing, all neurons remain active, and the expected output remains consistent with the behavior during training

# Regularization Experiment
As an experiment, we will train a pair of two-layer networks on 500 training examples: one will use no dropout, and one will use a keep probability of 0.25. We will then visualize the training and validation accuracies of the two networks over time.

```python
# Train two identical nets, one with dropout and one without.
np.random.seed(231)
num_train = 500
small_data = {
    'X_train': data['X_train'][:num_train],
    'y_train': data['y_train'][:num_train],
    'X_val': data['X_val'],
    'y_val': data['y_val'],
}

solvers = {}
dropout_choices = [1, 0.25]
for dropout_keep_ratio in dropout_choices:
    model = FullyConnectedNet(
        [500],
        dropout=dropout_keep_ratio
    )
    print(dropout_keep_ratio)

    solver = Solver(
        model,
        small_data,
        num_epochs=25,
        batch_size=100,
        update_rule='adam',
        optim_config={'learning_rate': 5e-4,},
        verbose=True,
        print_every=100
    )
    solver.train()
    solvers[dropout_keep_ratio] = solver
    print()
```

```python
# Plot train and validation accuracies of the two models.
train_accs = []
val_accs = []
for dropout_keep_ratio in dropout_choices:
    solver = solvers[dropout_keep_ratio]
    train_accs.append(solver.train_acc_history[-1])
    val_accs.append(solver.val_acc_history[-1])

plt.subplot(3, 1, 1)
for dropout_keep_ratio in dropout_choices:
    plt.plot(
        solvers[dropout_keep_ratio].train_acc_history, 'o', label='%.2f dropout_keep_ratio' % dropout_keep_ratio)
plt.title('Train accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(ncol=2, loc='lower right')

plt.subplot(3, 1, 2)
for dropout_keep_ratio in dropout_choices:
    plt.plot(
        solvers[dropout_keep_ratio].val_acc_history, 'o', label='%.2f dropout_keep_ratio' % dropout_keep_ratio)
plt.title('Val accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(ncol=2, loc='lower right')

plt.gcf().set_size_inches(15, 15)
plt.show()
```

## Inline Question 2:
Compare the validation and training accuracies with and without dropout -- what do your results suggest about dropout as a regularizer?

## Answer:
Comparing the validation and training accuracies with and without dropout provides insights into dropout as a regularization technique:

The results indicate that the model is prone to overfitting the training data, meaning it performs exceptionally well on the training data but struggles to generalize to unseen data.

__Without Dropout:__

In the absence of dropout, there is a significant disparity between the training accuracy (99.8% at epoch 25) and the validation accuracy (30.5% at epoch 25). This significant gap between training and validation accuracy suggests overfitting. Overfitting occurs when the model becomes too complex and fits the training data noise, leading to poor generalization.

__With Dropout:__

When dropout is applied, the training accuracy reduces slightly (to 92.2% at epoch 25), but the validation accuracy improves (to 32.4% at epoch 25). This means that dropout has a regularizing effect on the model. It bridges the gap between training and validation accuracy, which arises due to overfitting, by making the model more robust and less prone to fitting noise in the training data.

In summary, the results suggest that dropout acts as an effective regularizer. It helps control model complexity, preventing overfitting by encouraging the model to learn a simpler representation of the data. While it may lead to a slight drop in training performance, it significantly improves the model's ability to generalize to unseen data.







## Inline Question 3:
Suppose we are training a deep fully connected network for image classification, with dropout after hidden layers (parameterized by keep probability `p`). If we are concerned about overfitting, how should we modify `p` (if at all) when we decide to decrease the size of the hidden layers (that is, the number of nodes in each layer)?

## Answer:

If we decide to decrease the size of the hidden layers in a deep fully connected network that uses dropout as a regularization technique, and if we  are concerned about overfitting, we should consider increasing the value of the keep probability "p."

__Here's why:__

__Larger Hidden Layers:__ When we reduce the size of the hidden layers (i.e., the number of nodes in each layer), the network's capacity to model complex patterns in the data is reduced. Smaller hidden layers can potentially lead to underfitting, where the model struggles to capture the underlying patterns in the data.

__Dropout Effect:__ Dropout randomly deactivates a fraction of neurons during training, which can help prevent overfitting by making the network more robust. If we keep the dropout probability "p" too high (i.e., you drop too many neurons), it can limit the capacity of the network. With smaller hidden layers, we have fewer neurons to begin with, and a high dropout rate might result in too much information loss.

__Maintaining Balance:__ To maintain a balance between regularization and network capacity, we should consider increasing the value of "p" when reducing the size of the hidden layers. This means keeping more neurons active during training, which can help the network retain more information and make it more expressive.
