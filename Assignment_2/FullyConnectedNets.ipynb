Did you notice anything about the comparative difficulty of training the three-layer network vs. training the five-layer network? In particular, based on your experience, which network seemed more sensitive to the initialization scale? Why do you think that is the case?

## Answer:
When we contrast the challenge of training a three-layer neural network with that of training a five-layer neural network, it becomes evident that the latter is considerably more susceptible to variations in weight initialization scale. This heightened sensitivity can be attributed to the fact that the five-layer network is 66% deeper than the three-layer network. In deep neural networks, a small weight scale can result in vanishing gradients, characterized by exceedingly small products during backpropagation, while a large weight scale may lead to exploding gradients, indicated by excessively large products during the training process.

It's important to note that as the network's depth increases, the likelihood of encountering vanishing gradient issues becomes more prominent. Consequently, determining the optimal weight initialization scale for the five-layer network is a more intricate task compared to the three-layer network.

## Inline Question 2:

AdaGrad, like Adam, is a per-parameter optimization method that uses the following update rule:

```
cache += dw**2
w += - learning_rate * dw / (np.sqrt(cache) + eps)
```

John notices that when he was training a network with AdaGrad that the updates became very small, and that his network was learning slowly. Using your knowledge of the AdaGrad update rule, why do you think the updates would become very small? Would Adam have the same issue?


## Answer:
- **AdaGrad:** In AdaGrad, each update to the model's parameters is scaled down by the inverse square root of the accumulated sum of squared gradients. In simpler terms, it keeps track of how much each parameter has been updated over time. As training progresses, the accumulated sum of squared gradients can become quite large because it continuously adds positive values. Consequently, when you divide the current gradient by the square root of this large accumulated sum, the update becomes very small. This results in slow learning because the model doesn't adapt quickly

- **Adam:** Adam, on the other hand, Adam avoids this problem by employing an exponentially weighted moving average (EWMA) of gradients and their squares. In essence, it gives more importance to recent gradients while gradually reducing the influence of older ones. By doing this, Adam ensures that the EWMA of squared gradients remains reasonably limited in magnitude. Instead of dividing the gradient by the square root of the sum of all past squared gradients (as AdaGrad does), Adam divides the EWMA of the gradients (first moment) by the square root of the EWMA of the squared gradients (second moment). This approach allows Adam to control updates effectively, preventing them from becoming too large and ensuring faster convergence during training.


While both AdaGrad and Adam adapt the learning rates based on past gradient information, Adam's use of an EWMA approach helps it manage the learning rate more effectively, making it a popular choice for training neural networks.
